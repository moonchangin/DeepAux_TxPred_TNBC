```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Define the file path to the modified pickle file
file_path = '../alpha_output.pkl'

# Load the pickle file content
with open(file_path, 'rb') as file:
    alpha_data = pickle.load(file)

# Convert the list of arrays to a numpy array for easier indexing
alpha_data = np.array(alpha_data)

# Define the labels and colors for each arm
labels = ['Arm B', 'Arm C', 'Arm A - Negative Control', 'Arm A - Positive Control']
colors = ['red', 'blue', 'black', 'green']

# Plot the scatter-line plot
plt.figure(figsize=(10, 6))

for i in range(alpha_data.shape[1]):
    values = alpha_data[:, i]
    plt.plot(range(1, len(values) + 1), values, marker='o', linestyle='-', color=colors[i], label=labels[i])

# Add labels and title
plt.xlabel('Epochs')
plt.ylabel('Alpha')
plt.title('Scatter-Line Plot of Alpha Values Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load alpha values from the pickle file
with open('../27_alpha_output.pkl', 'rb') as f:
    alpha_data = pickle.load(f)

# Convert the list of numpy arrays to a 2D numpy array
alpha_array = np.array(alpha_data)

# Get the alpha values of the final epoch
final_epoch_alphas = alpha_array[-1]

# Identify the top 3 alpha values
top_3_indices = np.argsort(final_epoch_alphas)[-3:][::-1]
top_3_alphas = final_epoch_alphas[top_3_indices]

# Print the results
print("Top 3 alpha values and their indices:")
for idx, alpha in zip(top_3_indices, top_3_alphas):
    print(f"Index: {idx}, Alpha Value: {alpha}")

# Optionally, plot the alpha values
plt.figure(figsize=(10, 6))
for i, alpha_values in enumerate(alpha_array.T):
    plt.plot(alpha_values, label=f'Alpha {i}', linestyle='--' if i in top_3_indices else '-')
plt.xlabel('Epochs')
plt.ylabel('Alpha')
plt.title('Alpha Values Over Epochs')
plt.legend()
plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load alpha values from the pickle file
with open('../27_alpha_output.pkl', 'rb') as f:
    alpha_data = pickle.load(f)

# Convert the list of numpy arrays to a 2D numpy array
alpha_array = np.array(alpha_data)

# Get the alpha values of the final epoch
final_epoch_alphas = alpha_array[-1]

# Identify the top 3 alpha values
top_3_indices = np.argsort(final_epoch_alphas)[-3:][::-1]
top_3_alphas = final_epoch_alphas[top_3_indices]

# Print the results
print("Top 3 alpha values and their indices:")
for idx, alpha in zip(top_3_indices, top_3_alphas):
    print(f"Index: {idx}, Alpha Value: {alpha}")

# Define custom labels and colors
custom_labels = {
    1: 'GSE164458_veliparib_carboplatin_paclitaxel (main task)',
    13: 'GSE164458_paclitaxel (parallel arm task)',
    7: 'GSE192341_carboplatin (other parallel task)'
}
custom_colors = {
    1: 'red',
    13: 'blue',
    7: 'green'
}

# Plot the alpha values
plt.figure(figsize=(10, 6))
for i, alpha_values in enumerate(alpha_array.T):
    if i in custom_labels:
        plt.plot(alpha_values, label=custom_labels[i], color=custom_colors[i], linestyle='-')
    else:
        plt.plot(alpha_values, color='gray', linestyle='-', alpha=0.5)

# Set plot attributes
plt.xlabel('Epochs')
plt.ylabel('Alpha')
plt.title('Alpha Values Over Epochs')
plt.legend()
plt.show()
```
